## train/test split

```python
def train_test_split(dataset_df, split_date):
    train_df = dataset_df[dataset_df['date'] < split_date]
    test_df = dataset_df[dataset_df['date'] >= split_date]
    return train_df, test_df

train_df, test_df = train_test_split(top_50_areas_df, '2023-04-01')
```

```python
airport_df['last_week'] = airport_df['daily_trips'].shift(7)
airport_df = airport_df.dropna()
airport_df.iloc[:10]
```

```python
def mape(real, pred):
    return (abs(pred - real) / real).mean()


mape_day_last_week = mape(airport_df['daily_trips'], airport_df['last_week'])
mape_day_last_week
```

```python
airport_df['last_day'] = airport_df['daily_trips'].shift(1)
airport_df = airport_df.dropna()
airport_df.iloc[:10]
```

```python
mape_day_yesterday = mape(airport_df['daily_trips'], airport_df['last_day'])
mape_day_yesterday
```

```python
(mape_day_yesterday-mape_day_last_week) / mape_day_last_week
```

```python
airport_df['daily_error'] = (
    (abs(airport_df['last_day'] - airport_df['daily_trips']) 
     / airport_df['daily_trips'])
    *100
).astype(int)
airport_df
```

```python
airport_df['weekly_error'] = (
    (abs(airport_df['last_week'] - airport_df['daily_trips']) 
     / airport_df['daily_trips'])
    *100
).astype(int)
airport_df
```

```python
airport_df.groupby('Day_of_Week')['daily_error'].mean()
```

```python
airport_df.groupby('Day_of_Week')['weekly_error'].mean()
```

```python
airport_df['last_8_day'] = airport_df['daily_trips'].shift(8)
airport_df = airport_df.dropna()
airport_df.iloc[:10]
```

```python
airport_df['ratio'] = airport_df['last_day'] / airport_df['last_8_day']
airport_df
```

```python
airport_df['pred'] = airport_df['last_week'] * airport_df['ratio']
airport_df
```

```python
mape(airport_df['daily_trips'], airport_df['pred'])
```

```python
(mape_day_last_week-0.07613781985404944) / mape_day_last_week
```

```python
airport_df['pred_error'] = (
    ((airport_df['pred'] - airport_df['daily_trips']) 
     / airport_df['daily_trips'])
    *100
).astype(int)
airport_df
```

```python
airport_pred_error = airport_df.groupby('pred_error')['pred_error'].count()
airport_pred_error
```
